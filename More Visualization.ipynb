```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Using Global API

```python
x=np.arange(-10,11)
```

```python
plt.figure(figsize=(12,6))
plt.title('My First project using plot')
plt.plot(x,x**2)
plt.plot(x,-1*x**2)
```

```python
plt.figure(figsize=(12,6))
plt.title('My First project using plot')

plt.subplot(1,2,1)  #rows,columns & panel selected.
plt.plot(x , x**2)
plt.plot([0,0,0],[-10,0,100])
plt.legend(['x^2','Vertical Line'])
plt.xlabel('X')
plt.ylabel('X Squared')

plt.subplot(1,2,2)
plt.plot(x,-1*x**2)
plt.plot([-10,0,10],[-50,-50,-50])
plt.legend(['-x^2','Horizontal Line'])


plt.xlabel('X')
plt.ylabel('X Squared')
```

## OOP Interface

```python
fig,axes=plt.subplots(figsize=(12,6))
```

```python
axes.plot(
    x,(x ** 2) , color='red', linewitdth=3,
    marker='o', markersize=8 , label='X^2')

axes.plot(x,-1*(x**2),'b--',label='-X^2')

axes.set_xlabel('X')
axes.set_ylabel('X Squared')

axes.set_title("My First project using plot")

axes.legend()
fig
```
