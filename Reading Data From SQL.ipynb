# Reading data from relationsl databeses

```python
import pandas as pd
```

## Read data from SQL database

```python
import sqlite3
```

```python
conn=sqlite3.connect('chinook.db')
```

```python
cur=conn.cursor()
```

```python
cur.execute('SELECT*FROM employees LIMIT 5;')
```

```python
results=cur.fetchall()
```

```python
results
```

```python
df=pd.DataFrame(results)
```

```python
df.head()
```

```python
cur.close()
conn.close()
```

## Using pandas "read_sql" method

```python
conn=sqlite3.connect('chinook.db')
```

```python
df=pd.read_sql('SELECT*FROM employees;',conn)
```

```python
df.head()
```

```python
df=pd.read_sql('SELECT*FROM employees;',conn,index_col='EmployeeId',
               parse_dates=['BirthDate','HireDate'])
```

```python
df.head()
```

```python
df.info()
```

```python
df['ReportsTo'].mean()
```

```python
df['ReportsTo']>1.75
```

```python
df['City']=df['City'].astype('category')
```

```python
df.info()
```

## Using pandas "read_sql_query" method

```python
conn=sqlite3.connect('chinook.db')
```

```python
df=pd.read_sql_query('SELECT * FROM employees LIMIT 5;',conn)
```

```python
df.head()
```

```python
df=pd.read_sql_query('SELECT *FROM employees;',conn,index_col='EmployeeId',parse_dates=['BirthDate','HireDate'])
```

```python
df.head()
```

## Using "read_sql_table" method

```python
from sqlalchemy import create_engine
```

```python
engine=create_engine('sqlite:///chinook.db')
connection=engine.connect()
```

```python
df=pd.read_sql_table('employees',con=connection)
```

```python
df.head()
```

```python
df=pd.read_sql_table('employees',con=connection,index_col='EmployeeId',
                     parse_dates=['BirthDate','HireDate'])
```

```python
df.head()
```

```python
connection.close()
```
